## Data cleaning

### Subtask:
Clean the data by handling outliers, removing duplicates, and converting data types.


**Reasoning**:
Remove the 'Unnamed: 0' column, handle outliers in 'Purchase Amount (USD)', remove duplicates, convert 'Age' to integer, and handle other necessary type conversions.



```python
# Remove the 'Unnamed: 0' column
if 'Unnamed: 0' in df.columns:
    df = df.drop('Unnamed: 0', axis=1)

# Handle outliers in 'Purchase Amount (USD)' using IQR method
Q1 = df['Purchase Amount (USD)'].quantile(0.25)
Q3 = df['Purchase Amount (USD)'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['Purchase Amount (USD)'] >= lower_bound) & (df['Purchase Amount (USD)'] <= upper_bound)]

# Remove duplicate rows
df = df.drop_duplicates()

# Convert 'Age' column to integer
try:
    df['Age'] = df['Age'].astype(int)
except ValueError as e:
    print(f"Error converting 'Age' to integer: {e}")
    # Handle the error, e.g., replace invalid values with NaN or a default value
    # For this example, I'll print the error and continue
    pass

# Convert 'Review Rating' column to float
try:
    df['Review Rating'] = df['Review Rating'].astype(float)
except ValueError as e:
    print(f"Error converting 'Review Rating' to float: {e}")
    pass

# Display first 5 rows and info of the cleaned DataFrame
display(df.head())
df.info()
```

## Data analysis

### Subtask:
Analyze the relationship between customer demographics and purchase behavior.


**Reasoning**:
Analyze the relationship between customer demographics and purchase behavior by calculating the average, median, and total purchase amounts for different age groups and genders, and identify popular product categories and payment methods across demographics.



```python
# Group data by age group and gender to analyze purchase behavior
purchase_behavior = df.groupby(['Age_Group', 'Gender'])['Purchase Amount (USD)'].agg(['mean', 'median', 'sum'])
display(purchase_behavior)

# Analyze popular product categories across demographics
category_popularity = df.groupby(['Age_Group', 'Gender', 'Category'])['Customer ID'].count().unstack()
display(category_popularity)

# Analyze preferred payment methods across demographics
payment_method_preference = df.groupby(['Age_Group', 'Gender', 'Payment Method'])['Customer ID'].count().unstack()
display(payment_method_preference)

# Explore trends in purchasing behavior across locations or seasons.
season_purchase = df.groupby(['Season', 'Location'])['Purchase Amount (USD)'].agg(['mean', 'median', 'sum'])
display(season_purchase)
```

## Data visualization

### Subtask:
Visualize key findings from the data analysis.


**Reasoning**:
Visualize the key findings using matplotlib, including bar charts, box plots, and a pie chart.  The top 10 locations with the highest purchase amounts will also be visualized using a bar chart.



```python
import matplotlib.pyplot as plt

# 1. Season vs. Purchase Amount
plt.figure(figsize=(8, 6))
season_purchase = df.groupby('Season')['Purchase Amount (USD)'].agg(['mean', 'std'])
season_purchase.plot(kind='bar', y='mean', yerr='std', legend=False, capsize=4, color='skyblue')
plt.title('Average Purchase Amount by Season')
plt.xlabel('Season')
plt.ylabel('Average Purchase Amount (USD)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()


# 2. Purchase Amount Distribution by Age Group
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['Age_Group'] == group]['Purchase Amount (USD)'] for group in df['Age_Group'].unique()],
            labels=df['Age_Group'].unique(), patch_artist=True, showfliers=False)
plt.title('Purchase Amount Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Purchase Amount (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# 3. Most Popular Product Categories
plt.figure(figsize=(12, 6))
category_counts = df['Category'].value_counts()
category_counts.plot(kind='bar', color='lightcoral')
plt.title('Most Popular Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# 4. Distribution of Preferred Payment Methods
plt.figure(figsize=(8, 8))
df['Payment Method'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Distribution of Preferred Payment Methods')
plt.ylabel('')
plt.tight_layout()
plt.show()


# 6. Top 10 locations with the highest total purchase amounts
top_10_locations = df.groupby('Location')['Purchase Amount (USD)'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
top_10_locations.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Locations by Total Purchase Amount')
plt.xlabel('Location')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
